## Limpeza de Dados: Dados duplicados
### `.duplicated()` e `.drop_duplicates()`

O `.duplicated()` retorna uma máscara series booleana indicando se determinada linha possui duplicados ou não. Já o `.drop_duplicates()`, elimina as linhas duplicadas. <br>
Parâmetros do `.drop_duplicates()`:

   * `subset` -  seleciona colunas para serem utilizadas na comparação de linhas duplicadas
   * `keep` - {‘first’, ‘last’} (default ‘first’) Determina qual duplicado manter
   * `inplace` - (default False) Se False retorna uma cópia do df com as alterações, se True faz as modificações no próprio df



```python
df_CRM.duplicated()
```

```python
df_CRM.duplicated().sum()
```

```python
df_CRM[df_CRM.duplicated()].sort_values(by=["InvoiceNo", 'StockCode'])
```

```python
df_CRM[(df_CRM["InvoiceNo"]=='536749') & (df_CRM["StockCode"]=='21415')]
```

```python
df_CRM[df_CRM.duplicated(subset=["InvoiceNo", 'StockCode'])].sort_values(by=["InvoiceNo", 'StockCode']).iloc[110:120]
```

```python
# df_CRM[~df_CRM.duplicated()]
df_CRM.drop_duplicates()
# df_CRM.drop_duplicates(inplace=True)
df_CRM = df_CRM.drop_duplicates()
```

## Dados Nulos

**Problema:**

Dados nulos podem ser problemáticos pois indicam a falta de informação, o que pode afetar a precisão dos modelos e análises. Exemplos de problemas associados são:

- **Distorsão de Estatísticas:** Dados nulos podem levar a cálculos incorretos de medidas estatísticas, como média e desvio padrão.
- **Comprometimento de Modelos:** Em modelos de Machine Learning, dados nulos podem causar erros ou levar a resultados imprecisos.
- **Perda de Informações:** A exclusão de linhas com dados nulos pode resultar na perda de dados valiosos.

#### Exemplo: Empresa de E-commerce - Análise de Churn (Cancelamento)
Imagine uma empresa de e-commerce que deseja entender por que os clientes cancelam suas assinaturas (churn). Eles têm uma base de dados de clientes e uma coluna chamada "motivo_cancelamento".

O Problema: A coluna "motivo_cancelamento" está com 70% dos campos nulos, porque muitos clientes cancelam pelo app sem preencher o formulário opcional.

A Ação Errada: O analista de dados, querendo uma base "limpa", decide deletar todas as linhas onde "motivo_cancelamento" é nulo (df.dropna()).

O Resultado (O Problema): 
- Viés de Seleção: A empresa ficou apenas com os clientes que escreveram um motivo (geralmente reclamações muito específicas ou raivosas), perdendo a representação dos clientes que saíram por motivos silenciosos (preço, concorrência, desinteresse).
- Análise Distorcida: A empresa concluiu que o problema era "qualidade do produto" (motivo mais comum nas respostas preenchidas), quando na verdade a maioria saiu por "preço alto" (motivo que não foi capturado).
- Prejuízo Financeiro: A empresa investiu em melhoria de produto, mas o churn continuou alto porque o problema real era o preço, que foi ocultado pela deleção dos dados nulos

#### Removendo dados nulos

Uma das possíveis soluções para tratamento de dados nulos é a sua completa remoção.

**Removendo colunas:**

Quando temos uma coluna com boa parte dos dados nulos (**Ex.:** `Description`), significa que não conseguiríamos tirar insights dessas colunas.

Além disso, modelos de machine learning e ferramentas de visualização de dados não precisam utilizar esses dados.

Portanto, podemos remover esses dados sem problemas.

##### `.isna() e isnull()`
retorna um df booleano indicando se existe um nulo naquela posição. `pd.isnull()` é um alias para `pd.isna()` dentro do python como podemos ver na <a href="https://github.com/pandas-dev/pandas/blob/0409521665bd436a10aea7e06336066bf07ff057/pandas/core/dtypes/missing.py#L109">documentação</a>. O pandas dataframe é baseado nos df do R, onde null e na tem sentidos distintos.

```python
# Verificação do total e da porcentagem de dados nulos:
df_CRM.isna().sum()
```

```python
100*df_CRM.isna().mean()
```

```python
# Apagar os dados de Description, por ser praticamente vazio nos dados
df_CRM.drop(columns=["Description"], inplace=True)
```

```python
100*df_CRM.isna().mean()
```

**Removendo linhas nulas com `dropna()`:**

Às vezes algumas colunas de interese possuem vazios e fazem muita falta para compreensão dos dados. Por exemplo, no campo `CustomerID`, não temos 25% dos dados.

Podemos neste caso, remover os registros vazios, pois não sabemos se são clientes ou não.

```python
#Verificação da porcentagem de dados nulos:
df_CRM.isna().mean()
```

#### Estratégias para preenchimento de dados nulos:

Existem diversas estratégias para preencher dados nulos, entre elas, pode-se citar:

1. **Preenchimento com média, mediana ou moda**
- **Aplicação:** Recomendado para dados contínuos (média, mediana) e categóricos (moda).
- **Prós:**
  - Preserva a forma geral do conjunto de dados.
  - Rápido e eficiente para grandes conjuntos de dados.
- **Contras:**
  - Pode ser enganoso se a distribuição dos dados for assimétrica ou tiver outliers significativos.
  - Não leva em conta a possível correlação entre as variáveis.

##### `.fillna()`

preeche os valores nulos com valores especificados dentro do método. Esses valores podem tanto ser uma constante quanto aplicar técnicas como o backfill e o forward fill.

Parâmetros úteis:

   * `method {‘bfill’,‘ffill’, None}`, default None <br>
       Preenche os nas propagando o último valor válido para frente (ffill) ou utiliza próxima observação válida para preencher os nulos (bfill).
   * `limit`, default None <br>
       Quantidade máxima de nulos sequenciais a ser preenchido. 
   * `axis` {0 para ‘index’, 1 para ‘colunas’} <br>
       Eixo no qual se quer preencher os nulos
   * `inplace` {booleano}, default False


```python
# Definir a média da coluna UnitPrice dos dados
media_preco = df_CRM["UnitPrice"].mean()
print(media_preco)

# Inputar dados nulos
df_CRM["UnitPrice"].fillna(media_preco, inplace=True)

df_CRM.isna().mean()
```

### Aprofundamento

2. **Preenchimento com Valores Anteriores ou Posteriores**
- **Aplicação:** Ideal para séries temporais ou dados onde a sequência é importante.
- Prós:
  - Mantém a sequência lógica nos dados, o que é crucial em séries temporais.
- Contras:
  - Pode introduzir viés se houver mudanças abruptas ou tendências nos dados.

```python
# Preenchendo dados faltantes na coluna UnitPrice com valores anteriores ('ffill')
df_CRM["UnitPrice"].fillna(method = "ffill")
```

```python
# Preenchendo dados faltantes na coluna UnitPrice com valores posteriores ('bfill')
df_CRM["UnitPrice"].fillna(method = "bfill")
```

3. **Métodos de Interpolação**
- **Aplicação:** Útil em séries temporais ou quando os dados seguem uma tendência.
- **Prós:**
  - Oferece um método mais sofisticado e adaptativo para estimar valores nulos.
- **Contras:**
  - Pode ser complexo de implementar e entender.

[Documentação](https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.Series.interpolate.html)

Principais parâmetros: `Series.interpolate(method='linear', order=None, axis=0, limit=None, inplace=False)`

```python
# preencher com dados interpolados (interpolação linear)
df_CRM["UnitPrice"].interpolate(method="linear")
```

```python
# preencher com dados interpolados (interpolação polinomial)
df_CRM["UnitPrice"].interpolate(method="polynomial", order=2)
```

### Representação de Valores Ausentes em Conjuntos de Dados

Valores ausentes podem aparecer de diferentes formas, por isso é importante utilizar um formato consistente e bem documentado para uma manipulação limpa dos dados.

As representações comuns incluem:

- Células em branco: Entradas vazias em planilhas ou arquivos CSV.
- Valores especiais: Placeholders como "NA", "NaN", "NULL", "", "  " ou números como -999.
- Códigos ou sinalizadores: Rótulos como "MISSING" ou "UNKNOWN" para indicar tipos específicos de dados ausentes.

## Aprofundamento: Dados Inputados Errados

### Problemática

Dados inseridos incorretamente podem levar a análises errôneas. Podem ser erros tipográficos, valores fora de escala, entre outros.

### Resolução com .apply

A função .apply do Pandas permite aplicar uma função a cada elemento de uma coluna, o que pode ser usado para corrigir dados errados.

Suponha que uma coluna `Quantity` tenha um valores negativos, o que é claramente um erro.<br>

Suponha também que quantidades fracionadas podem ser também arredondadas.

```python
df_CRM["Quantity"].describe()
```

```python
def corrigir_quantidade(quantidade):
    # Verificar se a quantidade é negativa
    if quantidade < 0:
        return quantidade * -1
    else:
        return np.round(quantidade,0)
```

```python
# Aplicando o método apply
df_CRM["Quantity"] = df_CRM["Quantity"].apply(corrigir_quantidade)
df_CRM["Quantity"].describe()
```
