### Gráfico de barras - [barplot()](https://seaborn.pydata.org/generated/seaborn.barplot.html)
Utilizado em análise com variáveis **categóricas no eixo x e numéricas em y**. 

Nesse tipo de gráfico, **cada categoria é representada como uma única barra sendo o tamanho da barra representado pelo valor numérico** que pode ser, por exemplo, uma contagem, uma média, o valor máximo de cada categoria, etc.

__O seaborn considera no eixo y como valor a média da variável numérica__. Usando apenas o matplotlib precisávamos primeiro fazer a agregação.

```python
import numpy as np
# bar plot de 'Survived' como variável categórica em x pela variável numérica 'Fare' em y
sns.barplot(data=df, x='Survived', y='Fare', estimator=lambda x: len(np.unique(x)))
```

Pelo gráfico concluímos que quem mais pagou para embarcar teve mais chances de sobreviver.

Podemos adicionar uma dimensão a mais no nosso gráfico atribuindo cores diferentes para 'Sex' através do argumento `hue`

O que representam as barras de erro?

[errorbar](https://seaborn.pydata.org/generated/seaborn.barplot.html)

>> Ela representa um **Intervalo de Confiança de 95%** (ci=95). Isso significa que, se você coletasse os dados do Titanic várias vezes, a média real da tarifa estaria dentro daquele intervalo em 95% das vezes.

- **Barra maior:** Indica que os valores daquela categoria são muito variados (alta dispersão) ou que o tamanho da amostra é pequeno. Repare que para quem sobreviveu (Survived=1), a barra é maior, indicando que houve pessoas pagando preços muito diferentes (desde tarifas baratas até as caríssimas da 1ª classe).

- **Barra menor:** Indica que os valores estão mais "agrupados" em torno da média ou que a estimativa é mais precisa.

```python
stats = df.groupby(['Survived', "Sex"])['Fare'].agg(['mean', 'sem', 'std'])

stats['ci95_hi'] = stats['mean'] + 1.96* stats['sem']
stats['ci95_lo'] = stats['mean'] - 1.96* stats['sem']

stats
```

Isso significa que temos 95% de confiança que a proporção de mulheres que sobreviveram está na faixa de 43.7 e 60.2 de Fare. Se nós pegássemos uma amostra diferente dessa população, em 95% das vezes a média de Fare das mulheres que sobreviveram cairia dentro dessa faixa. 


O valor '1.96' é o "Z-score" para 95% de confiança. Ele assume que a distribuição das médias das amostras segue uma curva normal (Teorema Central do Limite). Para amostras grandes (geralmente 'n > 30'), esse cálculo é o padrão ouro na estatística frequentista. Pode haver uma doferença entre esse cálculo e a forma como o Seaborn calcula as barras por padrão:

- Seaborn (errorbar='ci'): Ele usa, por padrão, uma técnica chamada Bootstrapping (reamostragem com reposição). Ele sorteia os dados milhares de vezes para calcular o intervalo.

- Nosso cálculo: É uma fórmula analítica.

Exemplo: Se a coluna Fare (tarifas) tiver muitos valores extremos (outliers), o cálculo analítico e o bootstrapping do Seaborn podem dar valores ligeiramente diferentes, pois o bootstrapping lida melhor com dados muito "tortos" (assimétricos). Outro exemplo é quando temos pequenas amostras (T-Distribution). Se o seu grupo (por exemplo, "Homens que Sobreviveram") tiver poucas pessoas, o ideal seria usar o t-score em vez de $1.96$.

Outras barras:

- Para remover o error bar use o parâmetro: `errorbar=None`

- Para mudar o error bar use o parâmetro: `errorbar=(ci,90)` ou `errorbar='std`

[Error bars hide information](https://www.data-to-viz.com/caveat/error_bar.html)

### [Box plot](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

Box plots mostram a distribuição de dados categóricos ao ordená-los e dividi-los em 4 grupos com a mesma quantidade de dados em cada. Esse método facilita a identificação visual de outliers. <br>

Elementos de um boxplot:
   * 1º quartil (Q1): 25% percentil, ou seja, 25% dos seus dados têm valor menor que esse <br>
   * 3º quartil (Q3): 75% percentil, ou seja, 75% dos seus dados têm valor menor que esse <br>
   * Intervalo Interquartílico (IQR): distância entre 3º e 1º quartil (IQR = Q3 − Q1) <br>
   * Mediana: representa o meio da distribuição (50% percentil) <br>
   * Mínimo dentro do intervalo: valor mínimo dentro do dataset excluindo as anomalias (Q1 − 1.5 × IQR) <br>
   * Máximo dentro do intervalo: valor máximo dentro do dataset excluindo as anomalias (Q3 + 1.5 × IQR) <br>
   * Outliers ou Anomalias: observações extremas do dataset, qualquer ponto fora das barras  <br>

#### Describe x Boxplot

- No describe(): Os valores de min e max são os valores reais absolutos contidos no seu dataset (o menor e o maior valor de todos).

- No Boxplot: Os "bigodes" não vão necessariamente até o mínimo e o máximo. Eles respeitam o cálculo de barreira para identificar outliers:

    - Limite Superior: $Q3 + 1.5 \times IQR$
    - Limite Inferior: $Q1 - 1.5 \times IQR$

### [Scatterplot / Dispersão](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

Esse tipo de gráfico mostra a relação entre duas variáveis numéricas ao colocar cada uma das variáveis em um dos eixos dos gráficos. Com ele é possível avaliarmos o grau de correlação entre as duas variáveis.

>> Correlação é uma medida estatística que indica o quanto duas variáveis "caminham juntas". Ela descreve a força e a direção da relação entre dois fenômenos numéricos.

Correlação não implica causalidade. O fato de duas variáveis variarem juntas não significa que uma causa a outra. Elas podem estar sendo influenciadas por uma terceira variável oculta ou ser apenas uma coincidência estatística.

### [Line plot](https://seaborn.pydata.org/generated/seaborn.lineplot.html)

O gráfico de linha mostra a evolução das variáveis numéricas. Muito útil quando queremos ver dados no tempo.

O navio atingiu o iceberg às 23:40 e afundou completamente às 02:20. Vamos criar uma distribuição onde a maioria das mortes ocorre próximo ao horário final.

Vamos usar dados fake para gerar o horário em que as pessoas faleceram

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


# Criar um intervalo de tempo: das 23:40 às 02:20 (160 minutos)
horarios = pd.date_range(start="1912-04-14 23:40", end="1912-04-15 02:20", freq="min")

# Criar uma distribuição fake: poucas mortes no início, muitas no final (exponencial)
n_minutos = len(horarios)
pesos = np.exp(np.linspace(0, 5, n_minutos)) 
pesos /= pesos.sum()

# Distribuir os mortos reais do Titanic (aprox 1500) nesses minutos
mortes_por_minuto = np.random.multinomial(1500, pesos)

# Criar o DataFrame para o gráfico
df_timeline = pd.DataFrame({
    'Horário': horarios,
    'Mortes': mortes_por_minuto
})

# Calcular mortes acumuladas para um gráfico de linha mais fluido
df_timeline['Mortes Acumuladas'] = df_timeline['Mortes'].cumsum()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df_timeline, x='Horário', y='Mortes', color='red')
plt.title('Estimativa de Mortes por Minuto (Dados Sintéticos)')
plt.xticks(rotation=45)
plt.show()
```

Boas práticas:

- ordene os dados antes de plotar
- evite usar o lineplot quando não há uma relação de ordem entre os dados
- evite utilizar eixo duplo, prefira separar as categorias por cores
- não coloque muitas linhas ao mesmo tempo, isso pode dispersar a atenção. Se precisar, agrupe as categorias menos importantes em um único gráfico denominado "outros".
- cuidado com a razão entre a largura e comprimento do gráfico, isso pode enviesar a análise [link - aspect ratio](https://www.data-to-viz.com/caveat/aspect_ratio.html)
- ainda há muita discussão sobre cortar o eixo y em um valor diferente de zero [link](https://www.data-to-viz.com/caveat/cut_y_axis.html)

Podemos utilizar o método `corr()` que calcula a correlação linear (por padrão, de <a href="https://pt.wikipedia.org/wiki/Coeficiente_de_correla%C3%A7%C3%A3o_de_Pearson">Pearson</a>) entre as variáveis contínuas do nosso DataFrame e retorna uma matriz numérica com os valores dessa correlação.

```python
sns.heatmap(df.corr());
```

Onde __1__ quer dizer **correlação perfeita**, e __zero nenhuma correlação linear__, e __-1 correlação negativa perfeita__ (quando uma aumenta, outra diminui).

Para facilitar a visualização, podemos mudar o esquema de cores para o coolwarm ou quente/frio e mostrar os valores:

### Subplots

Se quisessemos vários gráficos em uma mesma imagem podemos usar o método subplot.

```python
fig, axes = plt.subplots(2, 2, figsize=(15,6))
fig.suptitle('2 x 2 subplots', fontsize=20)
axes[0,0].text(0.45, 0.5, str(('[0,0]')), fontsize=16)
axes[0,1].text(0.45, 0.5, str(('[0,1]')), fontsize=16)
axes[1,0].text(0.45, 0.5, str(('[1,0]')), fontsize=16)
axes[1,1].text(0.45, 0.5, str(('[1,1]')), fontsize=16);
```

```python
# Vamos criar quartis para 'Fare' utilizando a função pd.qcut
df['Fare_Range'] = pd.qcut(df['Fare'], 4)
```

Essa função pd.qcut distribui os valores em quantidades aproximadamente iguais entre os bins. Podemos validar aplicando a função .value_counts() na coluna criada de ranges de Fare

```python
df['Fare_Range'].value_counts()
```
