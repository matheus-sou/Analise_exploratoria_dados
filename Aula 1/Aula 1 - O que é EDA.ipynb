### 1. Definição do Problema

É a fase de negócio. Antes de tocar no código, você deve entender qual pergunta deseja responder ou qual processo deseja otimizar.

- Pergunta-chave: "Qual métrica de sucesso (KPI) queremos atingir?"

-  Ação: Traduzir uma dor de negócio (ex: "estamos perdendo clientes") em um problema de dados (ex: "prever a probabilidade de churn").

### 2. Data acquisition

- Selecionar dados relacionados ao problema.
- Extrair esses dados das fontes (csv, excel, json, banco de dados, crawler, API,...)

### 3. Data preparation

É a fase de transformação dos dados brutos (raw) em informações úteis que serão consumidas por analistas. A preparação dos dados é principalmente a fase que antecede a análise e tem como objetivo garantir a qualidade dos dados.

Principais passos:
- converter dados não estruturados em estruturados
- organizar os dados
- combinar diferente bases de dados, integrando-as
- limpar os dados:
    - identificar valores nulos
    - valores faltantes, removendo ou imputando novos valores
    - Remover duplicados
    - detectar ouliers e decidir o que fazer com eles
    - identificar dados que não adicionam valor e removê-los do dataset. Ex: nome das pessoas, CPF,..
- criar novas features
- tratar features categóricas
- garantir que dados respeitam a LGPD
- validar os dados:
    - resolver valores conflitantes como idade negativa, e-mail sem "@", endereços e telefones que não existem
    - garantir que cada coluna do dataset esteja no formato padrão da empresa. Ex: datas, telefones, endereços...

É a etapa de "conhecer o terreno". Aqui você limpa e entende os dados.

### 4. Modelagem

Onde a "mágica" estatística acontece. Aqui o objetivo é escolher e treinar algoritmos de Machine Learning (Regressão, Classificação, Clusterização) para identificar padrões nos dados históricos.

### 5. Validação

Verificar se o modelo realmente funciona e se resolve o problema original. Para isso é preciso testar o modelo com dados que ele nunca viu e medir a performance (usando métricas como Acurácia, Precisão, RMSE, etc.) analisando se modelo atingiu a meta definida na Etapa 1.

### 6. Deploy (Implantação)

Colocar o modelo em produção para que ele possa ser usado por outras pessoas ou sistemas. Geralmente isso é feito transformando o modelo em uma API ou integrando a um dashboard de tomada de decisão.

### 7. Monitoramento

Um modelo de dados não é "eterno"; ele pode perder a precisão conforme o mundo muda (conceito de Drift). Com isso devemos acompanhar continuamente a performance do modelo em produção. Se os resultados começarem a cair, o ciclo recomeça para retreinar o modelo com novos dados.


Ao longo do curso vocês passarão aprenderão mais sobre essas etapas, no módulo atual focaremos na preparação e entendimento dos dados: o famoso EDA!

## Análise Exploratória de Dados (EDA)

## Introdução

A **Análise Exploratória de Dados (EDA - Exploratory Data Analysis)** é uma
etapa fundamental no processo de análise de dados. Ela envolve a
**exploração e o entendimento inicial dos dados antes de análises mais
avançadas**.

------------------------------------------------------------------------

## O que é EDA?

EDA é uma abordagem inicial que busca:

Nessa fase os dados são analisados com o foco em tirar insights dos dados e respoder questões de negócios. Envolve utilizar ferramentas estatísticas para identificar e sumarizar as informações mais importantes do dataset e apresentá-las utilizando tabelas e gráficos.

### Principais passos:

- entender a solução e quais variáveis podem afetar

- avaliar distribuição dos dados e possíveis tratamentos

- identificar quais variáveis podem ser as mais relevantes para o modelo

- analisar correlação das variáveis

- identificar categorias dos dados que não adicionam valor e removê-las do dataset.

- identificar potenciais gaps nos dados ou se são enviesados para um gênero, sexo, religiçao raça. 

- identificar se existe algum grupo sem quantidade estatisticamente significante nos dados. Por exemplo: pessoas de diferentes localidades, grupos de idades, gênero, etc.

## Análise de Dados ou ML?

Suponha o problema: Uma grande varejista notou que as vendas caíram 15% em uma região específica. A diretoria queria um modelo de Churn Prediction (previsão de cancelamento) para oferecer cupons personalizados aos clientes que estavam parando de comprar.

- Análise de Dados: Um analista cruzou os dados de vendas com os dados de logística e frete.

- Descoberta: Ao plotar um gráfico de barras por região, notou-se que o tempo médio de entrega para aquela zona específica subiu de 2 para 10 dias devido ao fechamento de um centro de distribuição parceiro.

- Resolução: Não era um problema de "comportamento do cliente" que precisava de previsão, era um problema operacional. Resolver a logística recuperou as vendas sem precisar de uma única linha de código de ML.

## ANSCOMBE'S QUARTET - A Importância da Análise Exploratória de Dados (EDA)

Este exemplo clássico demonstra por que a EDA é fundamental em ciência de dados.

Os 4 datasets que serão analisados têm características estatísticas PRATICAMENTE IDÊNTICAS:
- Mesma média em X e Y
- Mesmo desvio padrão em X e Y
- Mesma correlação entre X e Y
- Mesma reta de regressão linear

````python
import seaborn as sns

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset('anscombe')
df1 = df.iloc[:11, 1:]  # extract first dataset
df2 = df.iloc[11:22, 1:]  # second dataset
df3 = df.iloc[22:33, 1:]
df4 = df.iloc[33:, 1:]
````

````python
df1.describe()
````

````python
df2.describe()
````

````python
df3.describe()
````

````python
df4.describe()
````

## Análise visual

NO ENTANTO, quando visualizamos graficamente cada dataset, descobrimos que:

Dataset 1: Relação linear perfeita com alguns desvios aleatórios

Dataset 2: Relação claramente NÃO-linear (curvada)

Dataset 3: Relação linear perfeita com um outlier extremo influenciando a reta

Dataset 4: Sem relação entre X e Y, com um único outlier criando uma correlação artificial

````python
import matplotlib.pyplot as plt

sns.lmplot(x='x', y='y', col='dataset', hue='dataset', data=df,
           col_wrap=2, ci=None, palette='muted',
           scatter_kws={'s': 50, 'alpha': 1})
plt.show()
````

## Conclusão
Este exemplo ilustra um perigo crítico em ciência de dados: NUNCA confiar apenas em métricas 
estatísticas simples sem explorar visualmente os dados. 

Métricas (mean, std, correlação) são RESUMOS que podem OCULTAR padrões importantes:
- Distribuições anômalas
- Outliers problemáticos
- Relações não-lineares
- Dados que violam suposições do modelo

A EDA com gráficos e visualizações revela a VERDADEIRA ESTRUTURA dos dados, 
evitando decisões analíticas incorretas que métricas isoladas poderiam sugerir.

Por isso a EDA é uma fase OBRIGATÓRIA antes de qualquer modelagem!

## 1. Primeiros Passos: Conheça o Dataset

**Objetivo:** Entender o formato e a estrutura básica dos dados.

**Tarefas:**
* **Carregar o dataset:** (ex.: CSV, JSON, SQL).
* **Inspecionar o dataset com métodos básicos:**
    * `head()`, `tail()`: Visualize as primeiras e últimas linhas.
    * `info()`: Confira o tipo das colunas e o total de valores não nulos.
    * `shape`: Verifique o número de linhas e colunas.
* **Identificar duplicatas e remover, se necessário:**
    ```python
    df.drop_duplicates(inplace=True)
    ```

---

## 2. Verificação de Qualidade dos Dados

**Objetivo:** Garantir que os dados estão prontos para análise.

**Tarefas:**
* **Valores ausentes:**
    * Identifique colunas com valores nulos: `isnull().sum()`.
    * Escolha estratégias para lidar com nulos:
        * Preencher com média/mediana para valores numéricos.
        * Remover linhas ou colunas, se necessário.
* **Tipos de dados:**
    * Confirme se os tipos das colunas estão corretos.
    * Converta tipos de dados, se necessário (ex.: `pd.to_datetime()` para colunas de data).
* **Análise de unicidade:**
    * Identifique colunas com valores únicos ou categorias limitadas.

---

## 3. Estatísticas Descritivas

**Objetivo:** Obter uma visão geral das distribuições e medidas básicas.

**Tarefas:**
* **Gere estatísticas descritivas com `describe()`:**
    * Média, mediana, desvio padrão, mínimo e máximo.
* **Analise variáveis categóricas:**
    * Frequência de cada categoria com `value_counts()`.
* **Analise variáveis numéricas:**
    * Verifique amplitude e valores extremos.

---

## 4. Exploração Visual dos Dados

**Objetivo:** Identificar padrões, tendências e outliers.

**Tarefas:**
* **Distribuições:**
    * Use histogramas para variáveis numéricas.
    * *Exemplo:* `sns.histplot(df['coluna'], kde=True)`.
* **Outliers:**
    * Utilize boxplots para detectar valores atípicos.
    * *Exemplo:* `sns.boxplot(data=df, x='coluna')`.
* **Relações entre variáveis:**
    * Scatterplots para variáveis numéricas.
    * Gráficos de barras para variáveis categóricas.
* **Correlação:**
    * Gere uma matriz de correlação para variáveis numéricas:
    ```python
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    ```

---

## 5. Análise Temporal (Se Aplicável)

**Objetivo:** Compreender padrões e tendências ao longo do tempo.

**Tarefas:**
* **Conversão de data:** Converta colunas para o formato datetime:
    ```python
    df['data'] = pd.to_datetime(df['data'])
    ```
* **Gráficos de linha:** Crie visualizações para analisar tendências:
    ```python
    df.groupby('data')['coluna_interesse'].mean().plot()
    ```

---

## 6. Identificação de Relações Entre Variáveis

**Objetivo:** Entender como as variáveis se relacionam.

**Tarefas:**
* **Correlações lineares:** Analise com scatterplots e matriz de correlação.
* **Relações categóricas vs. numéricas:** Use boxplots para comparar categorias em relação a variáveis numéricas.
* **Relações temporais:** Identifique sazonalidades ou picos inesperados.
