### `.describe()`
`.describe(include='all')` descreve **estatísticas básicas** sobre todas as colunas, inclusive as que são objetos como strings e timestamp. Nesse caso, *top* representará o valor mais comum enquanto *freq* será a frequência em que apareceu esse valor.

Esse método é utilizado para gerar um resumo estatístico das colunas numéricas de um DataFrame. Por padrão, ele calcula automaticamente os três principais quartis (25%, 50% e 75%), que dividem os dados em quatro partes iguais

- 25% (Primeiro Quartil - Q1): Indica que 25% dos valores são menores ou iguais a esse número.
- 50% (Segundo Quartil ou Mediana): O valor central que divide o conjunto de dados ao meio.
- 75% (Terceiro Quartil - Q3): Indica que 75% dos valores estão abaixo desse número e 25% acima.

Se você precisar de outros quantis (como decis ou percentis específicos), pode usar o parâmetro percentiles: 

```python
# Exemplo para obter quartis personalizados e outros percentis
df.describe(percentiles=[.1, .25, .5, .75, .9])
```

Ao carregar o dataset do Titanic, você notará que o Pandas identifica as colunas `PassengerId` e `Survived` como
  números inteiros (`int64`). Tecnicamente, isso está correto (elas contêm apenas números), mas do ponto de vista de
  Análise de Dados, essa classificação é semanticamente errada:

1. `PassengerId` (Identificador Único)
   * O que o Pandas vê: Uma sequência numérica crescente (1, 2, 3...).
   * O que realmente é: Uma variável qualitativa nominal. Esse número serve apenas como um "rótulo" ou "nome" para
     o passageiro.
   * Por que o int é ruim aqui? Não faz sentido matemático tirar a média dos IDs, somá-los ou calcular o desvio
     padrão. Tratá-los como números pode gerar gráficos estatísticos sem sentido (como um histograma de IDs).

2. `Survived`
   * O que o Pandas vê: Apenas zeros (0) e uns (1).
   * O que realmente é: Uma variável qualitativa binária (categórica/booleana). 
       * 0 representa "Não sobreviveu"
       * 1 representa "Sobreviveu"
   * Por que o int é enganoso aqui? Embora possamos tirar a média para saber a taxa de sobrevivência, o dado em si
     não é uma contagem ou medida, mas sim uma classificação. Em modelos de Machine Learning, é fundamental
     garantir que o algoritmo entenda que isso é uma categoria (classe) e não um valor escalar.


> O Pandas infere tipos baseando-se no formato do dado, mas ele não conhece o contexto do negócio. Como analistas,
  nosso papel é converter essas colunas para tipos adequados, como object ou category, logo após a leitura dos
  dados e é ai que entra o `.astype()`:




### `.astype()`
Converte o tipo da coluna.

1. Tipos de Dados Numéricos

    - 'int', 'int32', 'int64': Converte para números inteiros.
    - 'float', 'float32', 'float64': Converte para números de ponto flutuante.
    - 'uint8' a 'uint64': Inteiros sem sinal (positivos). 

2. Tipos de Dados de Texto e Objeto

    - 'str' ou 'string': Converte para texto (string). Muito usado para colunas que foram lidas incorretamente.
    - 'object': Converte para o tipo objeto genérico do pandas. 

3. Tipos Lógicos

    - 'bool': Converte para booleano (True/False). 

4. Datas e Tempo

    - 'datetime64[ns]': Converte para data e hora.
    - 'timedelta64[ns]': Converte para diferenças de tempo. 

5. Tipos Específicos do Pandas (Alta Performance)

    - 'category': Converte para dados categóricos. Essencial para otimização de memória e melhoria de desempenho quando uma coluna tem poucos valores únicos repetidos

Se quisermos fazer filtros mais complexos (filtros compostos, em mais de uma coluna), podemos fazer **conjunções entre filtros**, utilizando os **operadores lógicos de conjunção**. Aqui é obrigatório o uso dos parênteses para que os filtros sejam aplicados corretamente.

Obs.: temos os seguintes operadores lógicos:

- `&`     - corresponde ao "and"
- `|`     - corresponde ao "or"
- `~`     - corresponde ao "not"

````python
(df['Fare'] >= 260) & (df['Sex'] == 'female')
````

````python
2+(2*3)
````

````python
# filtar df para Fare >= 260 e Sex == 'female'
df[(df['Fare'] >= 260) & (df['Sex'] == 'female') & (df['Age']>=23)]
````

````python
# filtar df para Fare >= 260 ou Sex = female
df[(df['Fare'] >= 260) | (df['Sex'] == 'female')]
````

````python
# Podemos criar um novo df diretamento do filtro
df_mulheres_ricas = df[(df["Fare"]>260) & (df["Sex"] == 'female')]
df_mulheres_ricas
````

## Atenção: 

> **O uso de parênteses ao combinar condições no Pandas (como `&`, `|`, `~`) é obrigatório.**

Isso acontece por causa da precedência de operadores na linguagem Python:

1. Os operadores bitwise (`&`, `|`) têm uma precedência maior do que os operadores de comparação (`>=`, `==`, `<`).

2. Sem os parênteses, o Python tentaria executar o `&` primeiro. No seu exemplo, ele tentaria fazer: `260 & df["Sex"]`, o que geraria um erro de tipo (`TypeError`), pois não faz sentido lógico.

O que acontece na prática:

* Com parênteses (Correto):

    `df[(df["Fare"] >= 260) & (df["Sex"] == 'female')]`
    1. Resolve (`df["Fare"] >= 260`) -> Gera uma Série de True/False.
    2. Resolve (`df["Sex"] == 'female'`) -> Gera outra Série de True/False.
    3. O `&` compara as duas Séries e retorna o filtro final.

* Sem parênteses (Erro):

    `df[df["Fare"] >= 260 & df["Sex"] == 'female']`

Dica de Senior: Mesmo quando você tem apenas uma condição, como df`[df["Age"] > 30]`, os parênteses não são obrigatórios, mas ao usar qualquer
operador de combinação (`&`, `|`, `~`), coloque sempre parênteses em cada condição individual para evitar erros.

````python
# df[df["Fare"] >= 260 & df["Sex"] == 'female']
````

### Aprofundamento: Outras formas de filtrar
#### `.query()`
Se utilizar aspas simples para escrever a query, dentro dela utilize aspas duplas e vice-versa.

````python
df.query("Fare > 260 and Sex=='female'")
df.query("Fare > 260 & Sex=='female'")
df.query('Fare > 260 & Sex=="female"')
````

O método query também permite usarmos uma variável do workspace atual:

````python
fare_threshold = 260
gender = 'female'
df.query("Fare > @fare_threshold & Sex==@gender")
````

````python
# df.query("Fare > @fare_threshold & Sex like @gender") 
````

#### `.isin()`
Usado para filtrar Strings e valores numéricos. Passamos os valores dentro de uma lista.

````python
df[df.Name == 'Newell, Miss. Madeleine']
````

````python
# Vamos filtrar o df para os nomes: ['Newell, Miss. Madeleine', 'Fleming, Miss. Margaret']
df[df.Name.isin(['Newell, Miss. Madeleine', 'Fleming, Miss. Margaret'])]
````

````python
# Agora suponha que queremos apenas os passageiros com idade 1 ou 2
df[df.Age.isin([1,15])]
````

#### `.str.contains()`
O `.str.contains()` busca um padrão de texto dentro das strings.

````python
# E se quisermos saber quais passageiros eram da mesma família com sobrenome 'Newell'?
df[df.Name.str.contains('Newell')]
````

````python
df[df.Name.str.contains('well')]
````

````python
# E se quisermos saber quais passageiros eram da família com sobrenome 'Newell' ou 'Fleming'?
df[df.Name.str.contains('Newell|Fleming')]
````

#### `.str.startswith()`
Já nesse método não estamos buscando um padrão em qualquer local da string, queremos apenas aqueles que começam com esse valor.

````python
df[df['Name'].str.startswith('Newell')]
````

````python
df[df['Name'].str.startswith('newell')]
````

````python
df[df['Name'].str.startswith('New')]
````
